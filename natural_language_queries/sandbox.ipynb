```python
from pybaseball import statcast_batter
import matplotlib.pyplot as plt
import seaborn as sns

# Pete Alonso's MLB Player ID
player_id = 624413  # Pete Alonso

# Define the date (example: April 10, 2024)
game_date = "2024-04-24"
game_date_end = "2024-10-24"

# Get Statcast data for Pete Alonso on this date
df = statcast_batter(start_dt=game_date, end_dt=game_date_end, player_id=player_id)

# Filter for only batted balls (exclude strikeouts, walks, etc.)
df = df[df['events'].notna()]
df
```

```python
# Flip x-coordinates for accurate plotting (Statcast uses an inverted x-axis)
df['hc_x'] = 250 - df['hc_x']  # 250 is an approximation of Statcast's center field

# Create scatterplot with event-based color coding
plt.figure(figsize=(10, 10))
sns.scatterplot(
    x=df['hc_x'], y=df['hc_y'], hue=df['events'], style=df['events'],
    palette='Set1', s=100, edgecolor="black", alpha=0.7
)

# Customize labels and title
plt.xlabel("Spray Chart X")
plt.ylabel("Spray Chart Y")
plt.title(f"Pete Alonso Spray Chart ({game_date})")

# Show legend and plot
plt.legend(title="Event")
plt.show()
```

```python
import duckdb
import pandas as pd
import calendar
from pybaseball import statcast, batting_stats_range, pitching_stats_range
```

```python
con = duckdb.connect('statcast_2024_full.db')
```

```python
def store_statcast_data(year):
    # Fetch data month by month to manage memory
    for month in range(1, 13):  # Adjust month range if needed
        # Get the correct last day of the month
        last_day = calendar.monthrange(year, month)[1]
        start_date = f"{year}-{month:02d}-01"
        end_date = f"{year}-{month:02d}-{last_day}"

        # Fetch data
        df = statcast(start_dt=start_date, end_dt=end_date)

        # If the DataFrame is empty, skip DuckDB operations
        if df.empty:
            print(f"No data found for {start_date} to {end_date}, skipping...")
            continue

        # Store in DuckDB
        con.execute("""
        CREATE TABLE IF NOT EXISTS statcast_data AS 
        SELECT * FROM df WHERE 1=0
        """)
        con.execute("INSERT INTO statcast_data SELECT * FROM df")

# Example usage
store_statcast_data(2024)
```

```python
def store_batting_data(year):
    # Fetch data month by month to manage memory
    for month in range(1, 13):  # Adjust month range if needed
        # Get the correct last day of the month
        last_day = calendar.monthrange(year, month)[1]
        start_date = f"{year}-{month:02d}-01"
        end_date = f"{year}-{month:02d}-{last_day}"

        # Fetch data
        try:    
            df = batting_stats_range(start_dt=start_date, end_dt=end_date)
        except:
            print(f'data for month {month} not found')
            continue

        # If the DataFrame is empty, skip DuckDB operations
        if df.empty:
            print(f"No data found for {start_date} to {end_date}, skipping...")
            continue

        # Store in DuckDB
        con.execute("""
        CREATE TABLE IF NOT EXISTS batting_data AS 
        SELECT * FROM df WHERE 1=0
        """)
        con.execute("INSERT INTO batting_data SELECT * FROM df")

# Example usage
store_batting_data(2024)
```

```python

# Define long context and question
context = f"""
You are an expert SQL analytist with a specialization in baseball analytics. 
You have access to a duckdb connection with a table called batting_data.

Here is a detialed description of each of the columns in  that DB {batting_data_desc}

And here are the first ten rows of that table {result.head(10)}
""" 

question = f"please generate a query to describe {input_q}"

# Structure the message history
messages = [
    {"role": "system", "content": "You are an AI assistant that provides helpful answers."},
    {"role": "user", "content": f"Here is some context: {context}"},
    {"role": "user", "content": question}
]

# Call OpenAI API with the latest method
response = client.chat.completions.create(
    model="gpt-4-turbo",  # Use a model that supports long context (e.g., GPT-4 Turbo)
    messages=messages,
    max_tokens=500  # Adjust as needed
)

# Print the response
print(response.choices[0].message.content)
```

```python
qq = """
SELECT
    Name,
    mlbID,
    SUM(G) AS Games_Played,
    SUM(PA) AS Total_PA,
    AVG(BA) AS Avg_BA,
    AVG(OBP) AS Avg_OBP,
    AVG(SLG) AS Avg_SLG,
    AVG(OPS) AS Avg_OPS
FROM
    batting_data
WHERE
    G > 18
GROUP BY
    Name,
    mlbID
ORDER BY
    Avg_OPS DESC
LIMIT 1;
"""

con.execute(qq).df()
```
